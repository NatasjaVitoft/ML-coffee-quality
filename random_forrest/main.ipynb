```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/merged_data_cleaned.csv")
df.head(10)
```

```python
df.columns
```

```python

columns_to_drop = ['Unnamed: 0', 'Farm.Name', 'Lot.Number', 'Mill', 'ICO.Number', 'Company', 'Owner', 'Region', 'Producer', 'In.Country.Partner', 'Bag.Weight', 'Owner.1', 'Variety', 'Certification.Body', 'Certification.Address', 'Certification.Contact', 'Color', 'unit_of_measurement', 'Processing.Method', 'Altitude']
df_without_columns = df.drop(columns=columns_to_drop)
```

```python

df_without_columns.columns
```

```python
df_without_columns.head(10)
```

```python
df_new_names = df_without_columns.rename(columns={
    
    'Species': 'Sort',
    'Country.of.Origin': 'Country',
    'Number.of.Bags': 'Bags_amount',
    'Harvest.Year': 'Harvest_year',
    'Grading.Date': 'Grading_date',
    'Processing.Method': 'Processing_method',
    'Category.One.Defects': 'Category_one_defects',
    'Category.Two.Defects': 'Category_two_defects',
    'Total.Cup.Points': 'Total_cup_points'
    
})
```

```python
df_new_names.head(10)
```

```python

df_new_names.isnull().sum()
```

```python
df_new_names.shape
```

```python
df_new_names['Country'].unique()
```

```python
def map_region(country):
    if country in ['Ethiopia', 'Kenya', 'Uganda', 'Rwanda', 'Burundi', 'Malawi', 'Tanzania, United Republic Of', 'Zambia', 'Cote d?Ivoire']:
        return 'Africa'
    elif country in ['Guatemala', 'Costa Rica', 'Honduras', 'El Salvador', 'Nicaragua', 'Panama']:
        return 'Central America'
    elif country in ['Brazil', 'Colombia', 'Peru', 'Ecuador']:
        return 'South America'
    elif country in ['Indonesia', 'Vietnam', 'India', 'China', 'Myanmar', 'Thailand', 'Philippines', 'Laos']:
        return 'Asia'
    elif country in ['Mexico', 'United States', 'United States (Hawaii)', 'United States (Puerto Rico)']:
        return 'North America'
    elif country in ['Papua New Guinea', 'Taiwan', 'Japan', 'Mauritius', 'Haiti']:
        return 'Islands / Pacific'
    elif country in ['NaN']:
        return 'Unknown'
    else:
        return 'Other'

df_new_names['Region'] = df_new_names['Country'].apply(map_region)
```

```python
df_new_names.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df_new_names['Region_Label'] = le.fit_transform(df_new_names['Region'].astype(str))
```

```python
df_new_names.head(10)
```

```python
df_new_names['altitude_low_meters'] = df_new_names['altitude_low_meters'].fillna(df['altitude_low_meters'].median())
df_new_names['altitude_high_meters'] = df_new_names['altitude_high_meters'].fillna(df['altitude_high_meters'].median())
df_new_names['altitude_mean_meters'] = df_new_names['altitude_mean_meters'].fillna(df['altitude_mean_meters'].median())
```

```python
df_new_names.isnull().sum()
```

```python

df_new_names['Harvest_year'] = pd.to_numeric(df_new_names['Harvest_year'], errors='coerce')
df_new_names['Harvest_year'] = df_new_names['Harvest_year'].fillna(df_new_names['Harvest_year'].median())
```

```python
df_new_names.isnull().sum()
```

```python
df_new_names = df_new_names.dropna(subset=['Country'])
df_new_names = df_new_names.dropna(subset=['Quakers'])
```

```python
df_new_names.isnull().sum()
```

```python
columns_to_drop = ['Grading_date', 'Country']
df_without_columns_new = df_new_names.drop(columns=columns_to_drop)
```

```python
df_without_columns_new.head(10)
```

```python
df_without_columns_new['Sort'].unique()
df_without_columns_new['Sort_type'] = df_without_columns_new['Sort'].map({'Arabica': 0, 'Robusta': 1})
```

```python
df_without_columns_new.head(10)
```

```python
columns_to_drop = ['Sort', 'Region', 'Expiration']
df_without_columns_new = df_without_columns_new.drop(columns=columns_to_drop)
```

```python
df_without_columns_new.head(10)
```

```python
df_without_columns_new.to_csv('cleaned_dataset.csv', index=False)
```
